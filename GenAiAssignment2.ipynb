```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")
```

```python
# Array of prompts
prompts = [
    "The future of education is",
    "Artificial Intelligence will",
    "India in 2050 will be"
]

def generate_text(prompt, top_k, temperature):
    inputs = tokenizer.encode(prompt, return_tensors="pt")

    output = model.generate(
        inputs,
        max_new_tokens=50,
        do_sample=True,
        top_k=top_k,
        temperature=temperature
    )

    return tokenizer.decode(output[0], skip_special_tokens=True)

# Run for all prompts
for prompt in prompts:
    print("\nPrompt:", prompt)
    result = generate_text(prompt, top_k=40, temperature=0.7)
    print("Output:", result)
```

```python
# Different values to test
temperatures = [0.7, 1.0, 1.3]
top_k_values = [20, 40, 80]

# Store results
results = []

for prompt in prompts:
    for temp in temperatures:
        for k in top_k_values:

            output = generate_text(prompt, top_k=k, temperature=temp)

            results.append({
                "prompt": prompt,
                "temperature": temp,
                "top_k": k,
                "output": output
            })

# Print results clearly
for r in results:
    print("\n----------------------------------")
    print("Prompt:", r["prompt"])
    print("Temperature:", r["temperature"])
    print("Top_k:", r["top_k"])
    print("Output:", r["output"])
```
